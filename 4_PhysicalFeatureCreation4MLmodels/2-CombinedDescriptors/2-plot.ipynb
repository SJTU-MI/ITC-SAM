```python
from rdkit import Chem
import pandas as pd
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

strname = 'RF'

TCfile_path = './data/'+strname+'.csv'
figdata = pd.read_csv(
    TCfile_path,
)
import matplotlib.pyplot as plt
import seaborn as sns

# 创建图形和轴
_, axes = plt.subplots(1, 1, sharey=True, figsize=(5, 4))
sns.set_style('ticks')

# 绘制箱型图
sns.violinplot(
    x="TopFolder",
    y="tR2",
    hue='type',
    data=figdata,
    split=False,
    palette="Set2", 
    fliersize=1.5,  
    width=1.0,     
)

axes.spines['bottom'].set_linewidth(4)
axes.spines['left'].set_linewidth(4)
axes.spines['top'].set_linewidth(4)
axes.spines['right'].set_linewidth(4)

plt.xlabel('')
plt.ylabel('')
plt.title(strname, fontsize=30, fontweight='bold')

plt.xticks(fontsize=20, fontweight='bold')

custom_labels = ['2', '3', '4', '5', '6'] 
x_positions = range(len(custom_labels))
plt.xticks(ticks=x_positions, labels=custom_labels, fontsize=20, fontweight='bold')  

plt.yticks(fontsize=20, fontweight='bold')

legend = plt.legend()
legend.remove()


plt.tight_layout()
plt.savefig(strname+".png", dpi=1000)
plt.show()  
```
