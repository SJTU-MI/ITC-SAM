```python
from rdkit import Chem
import pandas as pd
import numpy as np
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

TCfile_path = './MLdata.csv'
figdata = pd.read_csv(
    TCfile_path,
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

_, axes = plt.subplots(1, 1, sharey=True, figsize=(9, 5))
sns.set_style('ticks')

sns.boxplot(
    x="SubFolder",
    y="tR2",
    hue='TopFolder',
    data=figdata,
    palette="Set2",  
    fliersize=1.5,   
    width=0.9,      
    order = ['1_RF_cal','2_XG_cal','3_GBR_cal','5_DT_cal','4_MLP_cal','6_KRR_cal'],
)

plt.ylim(0.3, 0.9)
axes.spines['bottom'].set_linewidth(3)
axes.spines['left'].set_linewidth(3)
axes.spines['top'].set_linewidth(3)
axes.spines['right'].set_linewidth(3)
plt.ylabel('Accuracy $R^2$', fontsize=22, fontweight='bold')
plt.xlabel('ML Model', fontsize=22, fontweight='bold')
plt.xticks(fontsize=20, fontweight='bold')
custom_labels = ['RF-I', 'XGB-II', 'GB-III', 'DT-IV', 'MLP-V', 'KR-VI'] 
x_positions = range(len(custom_labels)) 
plt.xticks(ticks=x_positions, labels=custom_labels, fontsize=20, fontweight='bold')  
plt.yticks(fontsize=20, fontweight='bold')
plt.legend(prop={'size': 14})
plt.legend(prop={'size': 14}, loc='upper right', ncol=3, bbox_to_anchor=(1, 1))
handles, labels = axes.get_legend_handles_labels()
new_labels = ['MDInter-1', 'Molecule-2', 'Mol2vec-3', 'MACCS-4', 'Morgan-5', 'cMorgan-6']
plt.legend(handles, new_labels, loc='upper right', ncol=3, prop={'size': 14}, bbox_to_anchor=(1, 1) )
plt.legend(handles, new_labels, loc='upper right', ncol=3, prop={'size': 14}, bbox_to_anchor=(1, 1), handletextpad=0.5,columnspacing=0.5 )
plt.tight_layout()
plt.savefig("boxplot.png", dpi=1000)
plt.show()  
```
